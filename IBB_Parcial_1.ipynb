Importar librerias

```python
import numpy as np #Funciones matematicas
import pandas as pd #Comunicación con Excel
from sklearn.naive_bayes import GaussianNB #Funcion Gaussiana
from sklearn.neighbors  import KNeighborsClassifier
from sklearn.cluster import KMeans
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix #La metrica de medicion
import matplotlib.pyplot as plt #Gráficador de cosas
import seaborn as sns #Gráficador de cosas
```

Cargar la base de datos

```python
nxl='/content/4. BD_Parcial_L_int.xlsx'
XDB=pd.read_excel(nxl,sheet_name=0)
XDB=XDB.dropna() #Elimina celdas sin datos
XDB['Sex'] = XDB['Sex'].replace({'M':0,'F':1}) # para convertir la variable sex como una variable categorica (0 o 1)

#Seleccion de las variables de trabajo
XD=np.array(XDB[['White', 'HS GPA', 'SAT', 'GPA', 'College Parent']]) #Datos entrada
yd=np.array(XDB[['Grad']]) #Dato de salida
```

### **3. ¿Cuál es el grupo de clasificación que logró la mayor calificación en la preparatoria (HS GPA) y el promedio de la Universidad (GPA)?**

```python
XDB.columns
```

```python
mnb=GaussianNB() #Llamamos el modelo
mnb.fit(XD,yd) # Ajustar el modelo a los datos de entrada y salida
ydp=mnb.predict(XD) # Pronostico intramuestral

#Parametros del modelo por categoria - Graduado / NO graduado
u=mnb.theta_ #Se estiman las medias
sigma=np.sqrt(mnb.var_) #Se estiman las varianzas

#Las variaciones de las variables
LI=u-sigma
LS=u+sigma

print('La media de las variables por categoria:')
print (u)
print('El limite inferior de la variación de las variables por categoria es:')
print(LI)
print('El limite superior de la variación de las variables por categoria es:')
print (LS)

#Métrica de evaluación para ver cómo se comporto el modelo identificando Graduado - No graduado
cm=confusion_matrix(yd,ydp) #Matriz de confusion
```

NOTA:

El array de arriba corresponde para los 'Grad = 0'

El array de abajo corresponde para los 'Grad = 1'

```python
print(f'La media del HS GPA de los no graduados es: {np.round(u[0, 2], 4)}')
print(f'La media del HS GPA de los que si se graduaron es: {np.round(u[1, 2], 4)}')

print(f'La media del GPA de los no graduados es: {np.round(u[0, 4], 4)}')
print(f'La media del GPA de los que si se graduaron es: {np.round(u[1, 4], 4)}')
```

### **4.¿De acuerdo con los índices de exactitud, sensibilidad y especificidad, determinar cómo fue el comportamiento del modelo?**

```python
VN=cm[0,0] #Verdaderos negativos
FP=cm[0,1] #Falsos positivos
FN=cm[1,0] #Falsos negativos
VP=cm[1,1] #Verdaderos positivos

Exactitud=(VN+VP)/len(XD)
print('La Exactitud es:',Exactitud) #Comportamiento general

Sensibilidad=VP/(FN+VP)
print('La Sensibilidad es:',Sensibilidad) #Como se comporto frente a los positivos

Especificidad=VN/(VN+FP)
print('La Especificidad es:',Especificidad) #Como se comportaron frente a los negativos
```

**Respuesta**

* La Exactitud del modelo es: **0.867**

* La Sensibilidad del modelo es: **0.926**

* La Especificidad modelo es: **0.733**

El modelo Naive Bayes presentó un rendimiento favorable. Su exactitud alcanzó el 86,70%, lo que demuestra una fuerte capacidad para predecir de manera precisa los casos generales. Además, la sensibilidad llegó al 92,63%, lo que subraya su excelente habilidad para identificar a las personas que se graduarán dentro del tiempo previsto. Sin embargo, la especificidad fue del 73,37%, lo que indica que es algo menos preciso al clasificar a quienes no se graduarán en el plazo de cuatro años. En resumen, el modelo mantiene un buen balance en sus predicciones, aunque se inclina ligeramente hacia la correcta identificación de aquellos que completarán su grado a tiempo.

### **6. ¿Cuál fue el comportamiento que tuvo el modelo frente a los índices de exactitud, sensibilidad y especificidad, y comparar este comportamiento con el modelo Naive Bayes?**

```python
mar=DecisionTreeClassifier(criterion='gini',max_depth=4) #4 niveles o variables de entrada
mar.fit(XD,yd)
ydp=mar.predict(XD)

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(yd,ydp)
print(cm)

#Evaluamos el modelo
VN=cm [0,0]
FP=cm [0,1]
FN=cm[1,0]
VP=cm[1,1]

Ex=(VN+VP)/(VN+VP+FN+FP) #Que porcentaje de los datos prónostico correctamente
print('La exactitud alcanzada por el modelo es:',Ex)
Se= VP/(FN+VP)
print('La sensibilidad alcanzada por el modelo fue:',Se)
Es=VN/(VN+FP)
print('La especificidad alcanzada por el modelo fue:',Es)
```

**Respuesta**

El modelo de árbol de decisión mostró un rendimiento positivo, con una exactitud del 90,2%, lo que lo convierte en un modelo muy eficaz para la clasificación general de los casos. Su elevada sensibilidad del 99,85% refleja que casi todas las personas que se graduarán fueron correctamente identificadas. No obstante, la especificidad fue menor, alcanzando solo el 68,51%, lo que evidencia dificultades para identificar adecuadamente a quienes no se graduarán en el plazo previsto. En síntesis, el árbol de decisión es muy efectivo para predecir casos positivos, pero menos preciso en la clasificación de los negativos.

Al compararlo con el modelo Naive Bayes, el árbol de decisión presenta una mayor exactitud (90,2% frente a 87,6%) y una sensibilidad significativamente superior (99,86% frente a 89,44%), lo que lo hace excelente para predecir quién se graduará. Sin embargo, Naive Bayes tiene un mejor desempeño en términos de especificidad (82,86% frente a 68,51%), siendo más preciso al identificar a quienes no completarán su grado. En resumen, el árbol de decisión sobresale en la predicción de graduaciones, mientras que Naive Bayes ofrece un mayor equilibrio entre las dos clases.

### **9.Determinar los clusters que lograron la mayor calificación en la preparatoria (HS GPA) y el promedio de la calificación en la Universidad (GPA)**

```python
np.random.seed(42)  # Para que todos tengamos el mismo modelo

#Se procede con la seleccion de las variables de trabajo
XD=np.array(XDB.iloc[:,1:6]) #Datos entrada
yd=np.array(XDB.iloc[:,6]) #Este es el dato que quiero que el modelo pronostique

# Crear el modelo de KMeans
mkm = KMeans(n_clusters=5, init='random', random_state=42)
mkm.fit(XD,yd)  # Solo usas XD para entrenar
labels=mkm.labels_
# Características de cada cluster
clusters = mkm.cluster_centers_
print('Las características de cada cluster son:\n', clusters)

C00=len(np.where((labels==0)&(yd==1))[0])
```

```python
# Para el cluster 0
print(f'El HS GPA medio del cluster 0 es: {XD[mkm.labels_ == 0][: , 1].mean()}')
print(f'El GPA medio del cluster 0 es: {XD[mkm.labels_ == 0][: , 3].mean()}')


# Para el cluster 1
print(f'El HS GPA medio del cluster 1 es: {XD[mkm.labels_ == 1][: , 1].mean()}')
print(f'El GPA medio del cluster 1 es: {XD[mkm.labels_ == 1][: , 3].mean()}')


# Para el cluster 2
print(f'El HS GPA medio del cluster 2 es: {XD[mkm.labels_ == 2][: , 1].mean()}')
print(f'El GPA medio del cluster 2 es: {XD[mkm.labels_ == 2][: , 3].mean()}')


# Para el cluster 3
print(f'El HS GPA medio del cluster 3 es: {XD[mkm.labels_ == 3][: , 1].mean()}')
print(f'El GPA medio del cluster 3 es: {XD[mkm.labels_ == 3][: , 3].mean()}')


# Para el cluster 4
print(f'El HS GPA medio del cluster 4 es: {XD[mkm.labels_ == 4][: , 1].mean()}')
print(f'El GPA medio del cluster 4 es: {XD[mkm.labels_ == 4][: , 3].mean()}')
```

**Respuesta**

El clúster que obtuvo la calificación más alta tanto en la variable 'HS GPA' como en 'GPA' fue el clúster 2, con los siguientes valores:

- El HS GPA medio del cluster 2 es: 3.9621

- El GPA medio del cluster 2 es: 3.1629

No obstante, la especificidad del 62.66% indica que el modelo es menos capaz de reconocer a aquellos que no se graduarán dentro del plazo, lo que lleva a un mayor número de falsos positivos. En conjunto, el modelo es más eficiente al predecir quiénes se graduarán, aunque podría mejorar en la clasificación de aquellos que no lo lograrán.

En comparación con otros modelos, este presenta una exactitud inferior al árbol de decisión (90.2%) y al Naive Bayes (86.67%). Aunque destaca por su alta sensibilidad del 95.01%, superior a la de Naive Bayes (92.63%) y cercana a la del árbol de decisión (99.86%), su especificidad es más baja (62.82%) en comparación con Naive Bayes (73.37%) y el árbol de decisión (68.51%). Esto indica que, si bien el modelo es efectivo para identificar a los estudiantes que se graduarán, es menos preciso al clasificar a aquellos que no lo harán, en comparación con los otros dos modelos. El árbol de decisión es extremadamente eficaz en la detección de casos positivos, pero presenta menor precisión en los negativos, mientras que Naive Bayes logra un mayor equilibrio entre ambas clasificaciones, superando a este modelo en especificidad y ofreciendo un desempeño más balanceado.

### **11.Cuál es el clúster que posee la mayor cantidad de personas que van a graduarse e indicar el porcentaje de graduación.**

```python
# Número de individuos totales por cluster
print(f"Cluster 0: {len(np.where(mkm.labels_ == 0)[0])} personas")
print(f"Cluster 1: {len(np.where(mkm.labels_ == 1)[0])} personas")
print(f"Cluster 2: {len(np.where(mkm.labels_ == 2)[0])} personas")
print(f"Cluster 3: {len(np.where(mkm.labels_ == 3)[0])} personas")
print(f"Cluster 4: {len(np.where(mkm.labels_ == 4)[0])} personas")
```

```python
# Número de individuos que SI van a graduarse por cluster
print(f"Cluster 0: {len(np.where((mkm.labels_ == 0) & (yd == 1))[0])} personas")
print(f"Cluster 1: {len(np.where((mkm.labels_ == 1) & (yd == 1))[0])} personas")
print(f"Cluster 2: {len(np.where((mkm.labels_ == 2) & (yd == 1))[0])} personas")
print(f"Cluster 3: {len(np.where((mkm.labels_ == 3) & (yd == 1))[0])} personas")
print(f"Cluster 4: {len(np.where((mkm.labels_ == 4) & (yd == 1))[0])} personas")
```

```python
# Porcentajes en el cluster 0
c00 = len(np.where((mkm.labels_ == 0) & (yd == 0))[0])
c01 = len(np.where((mkm.labels_ == 0) & (yd == 1))[0])
print('El porcentaje de no graduados para el cluster 0 es:', c00 / (c00 + c01))
print('El porcentaje de si graduados para el cluster 0 es:', c01 / (c00 + c01))


# Porcentajes en el cluster 1
c10 = len(np.where((mkm.labels_ == 1) & (yd == 0))[0])
c11 = len(np.where((mkm.labels_ == 1) & (yd == 1))[0])
print('El porcentaje de no graduados para el cluster 1 es:', c10 / (c10 + c11))
print('El porcentaje de si graduados para el cluster 1 es:', c11 / (c10 + c11))

# Porcentajes en el cluster 2
c20 = len(np.where((mkm.labels_ == 2) & (yd == 0))[0])
c21 = len(np.where((mkm.labels_ == 2) & (yd == 1))[0])
print('El porcentaje de no graduados para el cluster 2 es:', c20 / (c20 + c21))
print('El porcentaje de si graduados para el cluster 2 es:', c21 / (c20 + c21))

# Porcentajes en el cluster 3
c30 = len(np.where((mkm.labels_ == 3) & (yd == 0))[0])
c31 = len(np.where((mkm.labels_ == 3) & (yd == 1))[0])
print('El porcentaje de no graduados para el cluster 3 es:', c30 / (c30 + c31))
print('El porcentaje de si graduados para el cluster 3 es:', c31 / (c30 + c31))


# Porcentajes en el cluster 4
c40 = len(np.where((mkm.labels_ == 4) & (yd == 0))[0])
c41 = len(np.where((mkm.labels_ == 4) & (yd == 1))[0])
print('El porcentaje de no graduados para el cluster 4 es:', c40 / (c40 + c41))
print('El porcentaje de si graduados para el cluster 4 es:', c41 / (c40 + c41))
```
